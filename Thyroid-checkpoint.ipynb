Pathology: Specific types of thyroid cancer as determined by pathology examination of bpsy s* amples.
Focality: Whether the cancer is unifocal (limited to one location) or multifocal (present in multle loc* ations).
Risk: The risk category of the cancer based on various factors, such as tumor size, extent of spread, and stologi* cal type.
T: Tumor classification based on its size and extent of invasion into nearby s* tructures.
N: Nodal classification indicating the involvement of l* ymph nodes.

Recurred: Indicates whether the cancer has recurred after intial treatment.

**Now we've got our data split into training and test sets,it's time to build a machine learning model . we'll train it(find the patterns) on the traing set**

**And we'll test it(use patterns) on the test set**

**we're going to try 9 different machine learning model**

4. DecisionTressClassifier

```python
# Put the models in a dictionary
models={"Logistic Regression":LogisticRegression(),
       "KNN":KNeighborsClassifier(),
       "Random Forest":RandomForestClassifier(),
       "Decision Tree":DecisionTreeClassifier(),
       "Gaussian":GaussianNB(),
       "Ada Boost":AdaBoostClassifier(),
       "Gradient Boost":GradientBoostingClassifier(),
       "XGB":XGBClassifier(),
       "Cat Boost":CatBoostClassifier()}

# Create a function to fit and score models
def fit_and_score(models,x_train,x_test,y_train,y_test):
    """
    Fits and evaluates given machine learning models
     models:a dict of different Scit-learn machine learning models
     x_train:traing data(no labels)
     x_test:testing data(no labels)
     y_train:traing labels
     y_test:test labels
     """
    # Set random seed
    np.random.seed(42)
    # Make a dictionary to keep model scores
    model_scores = {}
    # Loop through models
    for name, model in models.items():
        # Fit the model to the data
        model.fit(x_train,y_train)
        # Evaluate the model and append its score to model_scores
        model_scores[name]=model.score(x_test,y_test)
    return model_scores
```

```python
model_scores=fit_and_score(models=models,
                            x_train=x_train,
                            x_test=x_test,
                            y_train=y_train,
                            y_test=y_test)
model_scores
```

```python
#fit models and generate classification reports
for name, model in models.items():
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    report = classification_report(y_test, y_pred)
    print(f'{name} Classification Report:\n{report}\n')
```
